Analysis
----------
From the Data above we can conclude that:
    - Maximum temperature goes up as the latitude nears the equator
    - Cloudiness and latitude does not seem to have any association
    - Wind speed and latitude also do not have any association

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#URL for API Call
query_url = "http://api.openweathermap.org/data/2.5/weather?appid="+api_key+"&units=imperial"+"&q="


#all required dataseries as lists
checked_date=[]
citylist = []
country_code= []
lat= []
long= []
temp_max = []
humid= []
cloud= []
wind= []


print("-"*50+"\nBeginning Data Retrieval\n"+"-"*50)

#loop through the cities list for API requests for each city
counter = 0
for city in cities:
    counter += 1
    #print log of each API requests
    print(f"Processing record {counter} of {len(cities)} for the city\t|{city}")
    try:
        response = requests.get(query_url + city).json()
    #appending all required data to the corresponding list from the json file
        citylist.append(response["name"])
        country_code.append(response["sys"]["country"])  
        lat.append(response["coord"]["lat"])    
        long.append(response["coord"]["lon"])
        temp_max.append(response["main"]["temp_max"])   
        humid.append(response["main"]["humidity"])
        cloud.append(response["clouds"]["all"])    
        wind.append(response["wind"]["speed"])
        checked_date.append(response["dt"])
        
        #Homework says exactly 500 cities
        if counter == 500:
            break
    except:
        print(f"City not found. Skipping...")
        counter -=1
        
print("-"*50+"\nData Retrieval Complete\n"+"-"*50)
```

```python
#Creating a DataFrame of all the weather data
weather_data = pd.DataFrame({"Country Code": country_code,
                             "Date" : checked_date,
                            "City": citylist,
                            "Latitude": lat,
                            "Longitude": long,
                            "Max Temp" : temp_max,
                            "Humidity": humid,
                            "Cloud" : cloud,
                            "Wind Speed": wind})

weather_data.to_csv("Weather_Data.csv",index=False)
weather_data.head()
```

```python
# Build a scatter plot for each data type

# Latitude vs Max Temp
plt.scatter(weather_data["Latitude"], weather_data["Max Temp"], marker="o",color ="lightblue",edgecolors = 'black')

# Add chart labels
plt.title("City Latitude vs Max Temperature (04/01/2019)")
plt.ylabel("Temperature (F) ")
plt.xlabel("Latitude")
plt.grid()

# Save the figure
plt.savefig("Lat_vs_MaxTemp.png")
# Show plot
plt.show()
```

```python
# Latitude vs Humidity
plt.scatter(weather_data["Latitude"], weather_data["Humidity"], marker="o",color="lightblue",edgecolors = 'black')

# Add chart labels
plt.title("City Latitude vs Humidity (04/01/2019)")
plt.ylabel("Humidity")
plt.xlabel("Latitude")
plt.ylim(-2, 102)
plt.grid()


# Save the figure
plt.savefig("Lat_vs_Humidity.png")

# Show plot
plt.show()
```

```python
# Latitude vs Cloudiness
plt.scatter(weather_data["Latitude"], weather_data["Cloud"], marker="o",color="lightblue",edgecolors = 'black')

# Add chart labels
plt.title("City Latitude vs Cloudiness (04/01/2019)")
plt.ylabel("Cloudiness")
plt.xlabel("Latitude")
plt.grid()


# Save the figure
plt.savefig("Lat_vs_Cloudiness.png")

# Show plot
plt.show()
```

```python

# Latitude vs Wind Speed
plt.scatter(weather_data["Latitude"], weather_data["Wind Speed"], marker="o",color="lightblue",edgecolors = 'black')

# Add chart labels
plt.title("City Latitude vs Wind Speed (04/01/2019)")
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid()

# Save the figure
plt.savefig("Lat_vs_WindSpeed.png")

# Show plot
plt.show()
```

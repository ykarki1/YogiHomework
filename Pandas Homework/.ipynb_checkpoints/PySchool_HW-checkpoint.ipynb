```python
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
merged = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
merged.head(20)
```

```python

number_of_schools = merged["school_name"].nunique()
number_of_students = merged["student_name"].count()
list_of_schools = merged["school_name"].unique()
total_budget = sum(merged["budget"].unique())
average_math = merged["math_score"].mean()
average_reading = merged["reading_score"].mean()
overall_passing = (average_math+average_reading)/2

#created new data frame for scores above and equal to 70 in math_score and reading_score
above70math = (merged.loc[(merged["math_score"] >=70)])
above70reading = merged.loc[merged["reading_score"]>=70]
math_passing = above70math["math_score"].count()/number_of_students*100
reading_passing = above70reading["reading_score"].count()/number_of_students*100


summary_dict =           {"Total Schools" : [number_of_schools],
                           "Total Students" : [number_of_students],
                          "Total Budget":[total_budget],
                           "Average Math Score": [round(average_math,2)],
                           "Average Reading Score": [round(average_reading,2)],
                           "% Passing Math": [round(math_passing,2)],
                           "% Passing Reading": [round(reading_passing,2)],
                           "% Overall Passing Rate": [round(overall_passing,2)]}

# converted the dictionary into a dataframe and applied formatting for specific columns
#Had to do this because .map function didn't work
summary_df = pd.DataFrame(summary_dict)
summary_df['Total Budget'] = summary_df.apply(lambda x: "${:,.2f}".format(x['Total Budget']), axis=1)
summary_df['Total Students'] = summary_df.apply(lambda x: "{:,}".format(x['Total Students']), axis=1)
summary_df.head()
```

```python
#School Summary

#grouped by school_name
byschool = merged.set_index("school_name").groupby(["school_name"])


#list of school types
school_types = school_data.set_index("school_name")["type"]

#list of numbers of students on each school
students_per_school = byschool["Student ID"].count()

# school budget
school_budget = school_data.set_index("school_name")["budget"]
school_budget


per_student_budget = school_budget/students_per_school


#avg scores by each school
avg_math = byschool["math_score"].mean()
avg_read = byschool["reading_score"].mean()

# % passing scores
pass_math = merged[merged["math_score"] >= 70].groupby("school_name")["Student ID"].count()/students_per_school*100 
pass_read = merged[merged["reading_score"] >= 70].groupby("school_name")["Student ID"].count()/students_per_school*100
overall = (pass_math+pass_read)/2

school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": students_per_school,
    "Per Student Budget": per_student_budget,
    "Total School Budget": school_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "% Passing Math": pass_math,
    "% Passing Reading": pass_read,
    "Overall Passing Rate": overall
    })
        

school_summary_formatted = school_summary.style.format({"Total Students": "{:,}", 
                          "Per Student Budget": "${:.2f}", 
                          "Total School Budget": "${:,.2f}",
                          "Average Math Score": "{:.2f}", 
                          "Average Reading Score": "{:.2f}", 
                          "% Passing Math": "{:.2f}%", 
                          "% Passing Reading": "{:.2f}%", 
                          "Overall Passing Rate": "{:.2f}%"})
school_summary_formatted
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Use of conditionals to obtain data series for each grades
ninth_graders = merged[(merged["grade"] == "9th")]
tenth_graders = merged[(merged["grade"] == "10th")]
eleventh_graders = merged[(merged["grade"] == "11th")]
twelfth_graders = merged[(merged["grade"] == "12th")]

ninth_graders_scores = ninth_graders.groupby(["school_name"]).mean()["math_score"]
tenth_graders_scores = tenth_graders.groupby(["school_name"]).mean()["math_score"]
eleventh_graders_scores = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
twelfth_graders_scores = twelfth_graders.groupby(["school_name"]).mean()["math_score"]

# Combining data series
mathscores_by_grade = pd.DataFrame({"9th": ninth_graders_scores, "10th": tenth_graders_scores,
                                "11th": eleventh_graders_scores, "12th": twelfth_graders_scores})

mathscores_by_grade = mathscores_by_grade[["9th", "10th", "11th", "12th"]]
mathscores_by_grade.index.name = None

#display the data
mathscores_by_grade
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#Use of conditionals to obtain data series for each grades
ninth_graders = merged[(merged["grade"] == "9th")]
tenth_graders = merged[(merged["grade"] == "10th")]
eleventh_graders = merged[(merged["grade"] == "11th")]
twelfth_graders = merged[(merged["grade"] == "12th")]

ninth_graders_scores = ninth_graders.groupby(["school_name"]).mean()["reading_score"]
tenth_graders_scores = tenth_graders.groupby(["school_name"]).mean()["reading_score"]
eleventh_graders_scores = eleventh_graders.groupby(["school_name"]).mean()["reading_score"]
twelfth_graders_scores = twelfth_graders.groupby(["school_name"]).mean()["reading_score"]

# Combining data series
reading_scores_by_grade = pd.DataFrame({"9th": ninth_graders_scores, "10th": tenth_graders_scores,
                                "11th": eleventh_graders_scores, "12th": twelfth_graders_scores})

reading_scores_by_grade = reading_scores_by_grade[["9th", "10th", "11th", "12th"]]
reading_scores_by_grade.index.name = None

#display the data
reading_scores_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
#binning
school_summary["spend Ranges (Per Student)"] = pd.cut(per_student_budget, spending_bins, labels=group_names)

spend_math_scores = school_summary.groupby(["spend Ranges (Per Student)"]).mean()["Average Math Score"]
spend_reading_scores = school_summary.groupby(["spend Ranges (Per Student)"]).mean()["Average Reading Score"]
spend_passing_math = school_summary.groupby(["spend Ranges (Per Student)"]).mean()["% Passing Math"]
spend_passing_reading = school_summary.groupby(["spend Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_rate = (spend_passing_math + spend_passing_reading) / 2

spend_summary = pd.DataFrame({"Average Math Score" : spend_math_scores,
                                 "Average Reading Score": spend_reading_scores,
                                 "% Passing Math": spend_passing_math,
                                 "% Passing Reading": spend_passing_reading,
                                 "% Overall Passing Rate": overall_passing_rate})

spend_summary = spend_summary[["Average Math Score", 
                                "Average Reading Score", 
                                "% Passing Math", 
                                "% Passing Reading",
                                "% Overall Passing Rate"]]
spend_summary
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Sample bins. Feel free to create your own bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_summary["School Size"] = pd.cut(school_summary["Total Students"], size_bins, labels=group_names)

# Calculate the scores based on bins
size_math_scores = school_summary.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = school_summary.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = school_summary.groupby(["School Size"]).mean()["% Passing Math"]
size_passing_reading = school_summary.groupby(["School Size"]).mean()["% Passing Reading"]
overall_passing_rate = (size_passing_math + size_passing_reading) / 2

size_summary = pd.DataFrame({"Average Math Score" : size_math_scores,
                             "Average Reading Score": size_reading_scores,
                             "% Passing Math": size_passing_math,
                             "% Passing Reading": size_passing_reading,
                             "% Overall Passing Rate": overall_passing_rate})

size_summary = size_summary[["Average Math Score", 
                             "Average Reading Score", 
                             "% Passing Math", "% Passing Reading",
                             "% Overall Passing Rate"]]

# Display output
size_summary
```
